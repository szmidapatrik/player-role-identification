```python
import soccerdata as sd
import pandas as pd
# set pandas dispaly options
pd.set_option('display.max_columns', None)
```

```python
fbref = sd.FBref('ENG-Premier League',seasons=["2017-18"],no_cache=True)
```

```python
meta_cols = ["season","team","player","nation","pos","age"]
```

```python
pl_table_shooting = fbref.read_player_season_stats(stat_type='shooting')
pl_table_shooting = pl_table_shooting.reset_index()
pl_table_shooting.columns = [' '.join(col).strip() for col in pl_table_shooting.columns.values]
pl_table_shooting = pl_table_shooting[meta_cols+["Standard Gls","Standard Sh","Standard SoT%","Standard Sh/90","Standard SoT/90","Standard G/Sh","Standard G/SoT","Standard Dist",
                                                 "Standard FK","Standard PK","Standard PKatt","Expected xG","Expected npxG","Expected npxG/Sh","Expected G-xG","Expected np:G-xG"]]
pl_table_shooting.head()
```

```python
# Get premier leagure table - passing
pl_table_passing = fbref.read_player_season_stats(stat_type='passing')
pl_table_passing = pl_table_passing.reset_index()
pl_table_passing.columns = [' '.join(col).strip() for col in pl_table_passing.columns.values]
pl_table_passing = pl_table_passing[meta_cols+["Total Cmp","Total Att","Total Cmp%","Total TotDist","Total PrgDist","Short Cmp","Short Att","Short Cmp%","Medium Cmp","Medium Att","Medium Cmp%","Long Cmp","Long Att","Long Cmp%","Ast","xAG","Expected xA","Expected A-xAG","KP","1/3","PrgP","CrsPA"]]
pl_table_passing.head()
```

```python
pl_table_passing_types = fbref.read_player_season_stats(stat_type='passing_types')
pl_table_passing_types = pl_table_passing_types.reset_index()
pl_table_passing_types.columns = [' '.join(col).strip() for col in pl_table_passing_types.columns.values]
pl_table_passing_types = pl_table_passing_types[meta_cols+["Pass Types Live","Pass Types Dead","Pass Types FK","Pass Types TB","Pass Types Sw","Pass Types Crs","Pass Types CK","Corner Kicks In","Corner Kicks Out","Corner Kicks Str","Outcomes Cmp","Outcomes Off","Outcomes Blocks"]]
pl_table_passing_types.head()
```

```python
# Get premier league table - goal_shot_creation
pl_table_goal_shot_creation = fbref.read_player_season_stats(stat_type='goal_shot_creation')
pl_table_goal_shot_creation = pl_table_goal_shot_creation.reset_index()
pl_table_goal_shot_creation.columns = [' '.join(col).strip() for col in pl_table_goal_shot_creation.columns.values]
pl_table_goal_shot_creation = pl_table_goal_shot_creation[meta_cols+["SCA SCA90","SCA Types PassLive","SCA Types PassDead","SCA Types TO","SCA Types Sh","SCA Types Fld","SCA Types Def","GCA GCA90","GCA Types PassLive","GCA Types PassDead","GCA Types TO","GCA Types Sh","GCA Types Fld","GCA Types Def"]]
pl_table_goal_shot_creation.head()
```

```python
# Get premier league table - defense
pl_table_defense = fbref.read_player_season_stats(stat_type='defense')
pl_table_defense = pl_table_defense.reset_index()
pl_table_defense.columns = [' '.join(col).strip() for col in pl_table_defense.columns.values]
pl_table_defense = pl_table_defense[meta_cols+["Tackles Tkl","Tackles TklW","Tackles Def 3rd","Tackles Mid 3rd","Tackles Att 3rd",
                                               "Challenges Tkl","Challenges Att","Challenges Tkl%","Challenges Lost",
                                               "Blocks Blocks","Blocks Sh","Blocks Pass","Int","Tkl+Int","Clr","Err"]]
pl_table_defense.head()
```

```python
# Get premier league table - possession
pl_table_possession = fbref.read_player_season_stats(stat_type='possession')
pl_table_possession = pl_table_possession.reset_index()
pl_table_possession.columns = [' '.join(col).strip() for col in pl_table_possession.columns.values]
pl_table_possession = pl_table_possession[meta_cols+["Touches Touches","Touches Def Pen","Touches Def 3rd","Touches Mid 3rd","Touches Att 3rd",
                                                     "Touches Att Pen","Touches Live","Take-Ons Att","Take-Ons Succ","Take-Ons Succ%","Take-Ons Tkld","Take-Ons Tkld%",
                                                     "Carries Carries","Carries TotDist","Carries PrgDist","Carries PrgC","Carries 1/3","Carries CPA","Carries Mis","Carries Dis","Receiving Rec","Receiving PrgR"]]
pl_table_possession.head()
```

```python
# Get premier league table - playing_time
pl_table_playing_time = fbref.read_player_season_stats(stat_type='playing_time')
pl_table_playing_time = pl_table_playing_time.reset_index()
pl_table_playing_time.columns = [' '.join(col).strip() for col in pl_table_playing_time.columns.values]
pl_table_playing_time = pl_table_playing_time[meta_cols+["Playing Time Mn/MP","Playing Time Min%","Playing Time 90s",
                                                         "Team Success PPM","Team Success onG","Team Success onGA","Team Success +/-","Team Success +/-90",
                                                         "Team Success On-Off","Team Success (xG) onxG","Team Success (xG) onxGA","Team Success (xG) xG+/-","Team Success (xG) xG+/-90",
                                                         "Team Success (xG) On-Off"]]
pl_table_playing_time.head()
```

```python
# Get premier league table - misc
pl_table_misc = fbref.read_player_season_stats(stat_type='misc')
pl_table_misc = pl_table_misc.reset_index()
pl_table_misc.columns = [' '.join(col).strip() for col in pl_table_misc.columns.values]
pl_table_misc = pl_table_misc[meta_cols+["Performance CrdY","Performance CrdR","Performance 2CrdY","Performance Fld",
                                         "Performance Off","Performance Crs","Performance Int","Performance TklW","Performance PKwon","Performance PKcon",
                                         "Aerial Duels Won","Aerial Duels Lost","Aerial Duels Won%"]]
pl_table_misc.head()
```

```python
dfs = [pl_table_shooting, pl_table_passing, pl_table_passing_types, pl_table_goal_shot_creation, pl_table_defense, pl_table_possession, pl_table_playing_time, pl_table_misc]
pl_table_all = pd.concat([df.set_index(meta_cols) for df in dfs], axis=1, join="inner").reset_index()
pl_table_all.head()
```

```python
pl_table_all.to_csv("./data/FBref/pl_table_2017_18.csv", index=False)
```

```python
understat = sd.Understat(['ENG-Premier League'],seasons=['2017-2018'],no_cache=True)
understat_shots = understat.read_shot_events()
understat_shots = understat_shots.reset_index()
understat_shots.head()
```

```python
understat_shots.body_part.unique()  
```

```python
understat_aggr = understat_shots[["player","location_x","location_y","minute"]].groupby("player").mean().reset_index()
understat_aggr.head()
```

```python
understat_aggr.to_csv("./data/Understat/understat_aggr_2017_18.csv", index=False)
```
